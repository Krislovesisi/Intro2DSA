```python
import pandas as pd
import numpy as np

housingInfo = pd.read_csv('Tree-classification.csv')
housing = housingInfo.values

housingPrice=[]
for x in housing:
    if x[1] > 160:
        if x[3] > 3:
            if x[2] < 10:
                housingPrice.append(510)
            else: housingPrice.append(382)
        elif x[2] < 40:
            housingPrice.append(335)
        else: housingPrice.append(312)            
    elif x[2] < 25:
        if x[3] > 2:
             housingPrice.append(341)
        else: housingPrice.append(258)
    elif x[1] > 100:
                housingPrice.append(285)
    else: housingPrice.append(217)

housingInfo["housingPrice"]=housingPrice
print(housingInfo)

        
```
